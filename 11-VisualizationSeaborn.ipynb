```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull()
```

```python
data.info()
```

```python
plt.scatter("Height", "Weight", data= data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='Height',y='Weight', data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python

sns.set_style("whitegrid")   ## arkaplan rengini ve gridleri degistirebilirim. mesela darkgrid diyebilirim 
sns.scatterplot(x= 'Height', y='Weight', hue='Sex', data=data)     ### --> hue bize mesela cinsiyetleri ayri ayri olarak gormemize yardimci oldu.

plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')

plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex", style="Medal" , size="Age", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='Height', y='Weight', hue="Sex", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', hue="Sex",data=data)
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
#kde - kernel density estimate
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', hue="Sex",data=data, kind="kde")
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', y='Weight', hue="Sex",data=data, kind="kde")
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.barplot(x='Medal', y='Height', hue="Sex", data=data)
plt.title('Medals by Height')
plt.show()
```

```python
sns.catplot(x='Medal', y='Height', hue='Sex', col='Season', data=data, kind='bar')
```
