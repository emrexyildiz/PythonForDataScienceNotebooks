```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_lits = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_lits, "r")  #----> y yellow demek.
plt.title("Age VS Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
#numpy ile calismalar yapiyoruz. Numpy daha hizli calistigi icin numpy hesaplamalarida hizli yapar. C dili ile optimize edilmistir. 
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_lits)
```

```python
plt.plot(np_age_list, np_weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age VS Weight")
plt.show()
```

```python
#different examples
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
plt.plot(numpy_arr1, numpy_arr2)
plt.show()


```

```python
plt.plot(numpy_arr1, numpy_arr2, "g*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "r*-")
plt.subplot(1,2,2)                            # 1 row, 2 columns, 2nd graph
plt.plot(numpy_arr2, numpy_arr1, "g--")
plt.show()
```

```python
#customize the axes
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.6,0.6])     #burasi kutunun uzunluguna kisalgina verilen deger. 
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")

plt.show()
```

```python
#customize labels
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])  #kutuyu cizdik
new_axes.plot(numpy_arr1,numpy_arr1 **2, label = "numpy array ** 2")  #burada hem label isini hallettik hemde datayi girdik.
new_axes.plot(numpy_arr1,numpy_arr1 **3, label = "numpy array ** 3")
new_axes.legend(loc=2)   #burada ise label nereye istersek oraya yerlestirdik.

plt.show()
```
