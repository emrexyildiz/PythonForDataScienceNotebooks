```python
### simdi burada fonksiyon ile maaslari yeni bir column altinda tespit edicez. 50 binden az ise lOW yazdircez
### 50 ile 80 arasi ise salary MEDIUM yazdircaz
### 80 bin uzeri ise de HIGH yazdircaz.
```

```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary <= 80000:     #50 bin ile 80 bin arasi demek
        return "medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)     ### burada df Salary icine bu olusturdgumuz function baglamis durumdayiz. 
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)   ## burada df icine Salary_Category diye bir column olusturduk
                                                              ## ve fuction'a bagladik artik. 
```

```python
# Example 2: Apply a function to calculate adjusted performance score
# Burada eger Experience 10 yildan fazla ise performance score'a 1 ekle diyoruz. bi nevi reward yapiyoruz bu kisiyi.
```

```python
def adjust_performance(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    return row["Performance_Score"]
```

```python
df["Adjusted_Performance"] = df.apply(adjust_performance, axis=1)
```

```python
df
```

```python
# Example 3   -------> Bu defa experiance 5 yildan azsa -1 verdik performance score
```

```python
def adjust_performance_1(row):
    if row ["Experience"] < 5:
        return row ["Performance_Score"] -1
    else:
        return row ["Performance_Score"]
```

```python
df["Adjusted_Performance-1"] = df.apply(adjust_performance_1, axis=1)
```
