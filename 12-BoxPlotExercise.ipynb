```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# box plot
```

```python
# Understanding Box Plots
# Box plots (also called whisker plots) are used to visualize the distribution of a dataset.
# They provide insights into:
# - The median (Q2, 50th percentile)
# - The interquartile range (IQR) - spread between Q1 (25th percentile) and Q3 (75th percentile)
# - The minimum and maximum values within 1.5 * IQR (whiskers)
# - Outliers (values outside the whiskers)
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title('Box Plot of Simple Numreic Data')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
### Interpretation ###
# 1. The box represents the interquartile range (IQR), which contains 50% of the data.
# 2. The line inside the box is the median (50th percentile).
# 3. The whiskers extend to 1.5 times the IQR.
# 4. Any points beyond the whiskers are considered outliers.
# 5. Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).
```

```python
data= np.array ([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60,100,500])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)

plt.title("Boc Plot of Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(13,8))

plt.subplot(1,2,1)

sns.boxplot(x="class", y="age", data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
